```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model
import cv2
```

```python
model = load_model('C:\\Users\\gotze\\OneDrive\\Desktop\\ESD Mf\\Model2.keras')
```

```python
emotions = ['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise']
```

```python
def predict_emotion(model, img_path):
    img = cv2.imread(img_path)
    img = np.expand_dims(img, axis=0) / 255.0
    
    predictions = model.predict(img)
    max_index = np.argmax(predictions)
    
    return emotions[max_index]
```

```python
predict_emotion(model, 'C:\\Users\\gotze\\OneDrive\\Desktop\\ESD Mf\\images\\validation\\neutral\\1493.jpg')
```
